```python
import time
import json
import random

from sdamgia import SdamGIA

subject = 'mathb'  # наши предметы: bio, fr, inf, math, mathb, phys, sp
sdamgia = SdamGIA()

all_problems = []
counter = 0

# Получаем структуру каталога
catalog = sdamgia.get_catalog(subject)

for topic in catalog:
    topic_name = topic.get('topic_name', '')
    for category in topic['categories']:
        category_id = category['category_id']
        category_name = category.get('category_name', '')
        # Получаем список задач в категории
        problem_ids = sdamgia.get_category_by_id(subject, category_id)
        for problem_id in problem_ids:
            try:
                problem = sdamgia.get_problem_by_id(subject, problem_id)
                if problem:
                    # Добавляем метаинформацию о топике и категории
                    problem['topic_name'] = topic_name
                    problem['category_name'] = category_name
                    all_problems.append(problem)
                    counter += 1
                    print(f"Собрана задача {problem_id} ({counter} всего)")
                    # Сохраняем промежуточные результаты каждые 5 задач
                    if counter % 5 == 0:
                        with open(f'{subject}_problems_partial.json', 'w', encoding='utf-8') as f:
                            json.dump(all_problems, f, ensure_ascii=False, indent=2)
                else:
                    print(f"Пустой ответ для задачи {problem_id}")
            except Exception as e:
                print(f"Ошибка при обработке задачи {problem_id}: {e}")
            time.sleep(3 + random.uniform(0, 2))  # Задержка между запросами чтобы сайт не ругался

# Финальное сохранение всех задач
with open(f'{subject}_problems_full.json', 'w', encoding='utf-8') as f:
    json.dump(all_problems, f, ensure_ascii=False, indent=2)

print(f"Собрано всего задач: {counter}")
```
